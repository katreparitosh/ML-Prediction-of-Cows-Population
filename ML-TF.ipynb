```
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression as LinReg
```

```
#Cows 1.30 1.02 0.96 0.97 0.88 0.65 0.44 0.32
livestock_ = np.array([1.30, 1.02, 0.96, 0.97, 0.88, 0.65, 0.44, 0.32])
#Species 1951 1982 1987 1992 1997 2003 2007 2012
year_ = np.array([1951, 1982, 1987, 1992, 1997, 2003, 2007, 2012])
import matplotlib.pyplot as plt
```

```
df = pd.DataFrame({'Cows': livestock_,  'Year': year_    })
df
# Plot livestock and year 
plt.plot(df['Year'], df['Cows'], color='blue')  # bx = blue x
plt.xlabel("Year")
plt.ylabel("Cows Population in India (M)")
plt.title('Cows population - India Year Wise - Source : https://worldmapper.org/maps/donkeys-2016/')
plt.show()
```

```
x = year_.reshape(-1, 1)
#y = grouped['LandAverageTemperature'].values
#x = year_
y = livestock_

reg = LinReg()
reg.fit(x, y)
y_preds = reg.predict(x)
print("Accuracy: " + str(reg.score(x, y)))

plt.figure(figsize = (15, 5))
plt.title("Linear Regression")
plt.scatter(x = x, y = y_preds)
plt.scatter(x = x, y = y, c = "r")
plt.show()


prediction = reg.predict(2045)
print('Prediction for 2045 - Cows count in India :' + str(prediction) + ' Accuracy: 82.665184%')
```
